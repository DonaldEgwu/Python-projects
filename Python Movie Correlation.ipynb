```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib 

import matplotlib.pyplot as plt

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] =(12,8) #adjust the configuration it will create

#read the data

df =pd.read_csv(r"C:\Users\user\Desktop\data anayltics\Projects and Resume\Docs\movies.csv")
```

```python
# looking at the data

df.head(10)
```

```python
# looking for missing data

for col in df.columns:
    percent_missing= np.mean(df[col].isnull())
    print('{} - {}%' . format(col,percent_missing))
```

```python
# date types for columns

df.dtypes
```

```python
#changing data type of the columns budget and gross

df['budget']=df['budget'].astype('int64') # this actually gets rid of the decimal by coverting to int from float 

df['gross']=df['gross'].astype('int64')

df.head(10)
```

```python
#create correct year column
df['correct year']= df['released'].astype(str).str[:4]

df.head(10)
```

```python
#TO ORDER BY GROSS 
df=df.sort_values(by =['gross'],inplace =False,ascending =False) #'ascending= False' is to order by either ascending or descending 
```

```python
df_numerized = df
# using a for loop
#remenber the spacing is important
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
     df_numerized[col_name] = df_numerized[col_name].astype('category')
     df_numerized[col_name]  = df_numerized[col_name].cat.codes

df_numerized.head(10)
     
```

```python
df.head(10)
```

```python
corr_matrix =df_numerized.corr(method ='pearson')
sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation Matrix for numeric features')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features')

plt.show()
```

```python
df_numerized.corr() #remember this is the "pearson" method
```

```python
#to see the correlation more easily you use a method called "unstacking"

corr_matrix = df_numerized.corr()
corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
pd.set_option("display.max_rows",None)
```

```python
#to sort the matrix
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
#to sort by high correlation

high_corr = sorted_pairs[(sorted_pairs) >0.5]

high_corr
```
